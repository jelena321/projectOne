```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = "top50.csv"
data = pd.read_csv(filename, encoding='ISO-8859-1')
data.info()
```

```python
#prvi zadatak
top10longest = data.nlargest(10,["Length."])
top10longest
top10longest.to_csv("top10longest.csv")
```

```python
#drugi zadatak
danceability = data.nlargest(10,["Danceability"])
danceability
```

```python
#treci zadatak
new_song = []
for i in range (len(data["Track.Name"])):
    if data["Track.Name"][i][0].isupper():
        new_song.append(data.iloc[i]) #iloc uzima sve vrijednosti iz red na lokaciji [koja je zadata]
new_song1 = pd.DataFrame(data = new_song) #DataFrame pravi tabelu od liste
new_song1.nlargest(10,["Popularity"]).sort_values("Energy",ascending = False)
```

```python
#cetvrti zadatak
count = data["Genre"].value_counts()
plt.barh(count.index, count.values)
```

```python
#peti zadatak
new_col= []
tabela = pd.DataFrame()
for col in data.columns:
    if col not in ["Unnamed: 0","Track.Name", "Artist.Name", "Genre"]:
        new_col.append(col) 
for i in range (len(new_col)):
     tabela.insert(i, new_col[i], data[new_col[i]])
a = tabela.corr().replace(1,0).max().max()
a
```
